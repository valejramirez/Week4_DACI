```python
#Caso sin efectos
from lcapy import Circuit, t, pi
import sympy as sy
from sympy import diff

cct1= Circuit("""

P1 0_1 0; down=1, v=V{in}
Rs 0_1 1 ; up
W 1 1_1; right
G1 1_1 2 0 1  gm; up, l=gm Vgs  
W 2 2_1; right
Rd 2_1 0_2; down =3
W 0 0_2; right
W 2_1 2_2; right
W 0_2 0_3; right
P2 2_2 0_3;  v^=V{out}
""")
cct1.draw(node_spacing=2.5)
H = cct1.P1.transfer('P2')
H.simplify_terms()
```

```python
Zin=cct1.impedance('P1').simplify_terms()
Zin
```
